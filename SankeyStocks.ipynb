```python
from pandas.core.arrays import numeric
#Required parameters
url = "https://www.alphavantage.co/query?"
api_key="P1RE3MNTG259TT0B"
function="INCOME_STATEMENT"
symbol="IBM"

#requesting the data with the Alpha Vantage API
response=requests.get(f"{url}function={function}&symbol={symbol}&apikey={api_key}")

if response.status_code==200:
  is_data=response.json()

else:
  print(f"Error: {response.status_code} - {response.text}")



                                                                    ########## Annual Reports Code #############
#extracting annual reports from the json response file
annual_reports=is_data['annualReports']

#converting the json file into a Pandas DataFrame
annual_df=pd.DataFrame(annual_reports)

#converting the numeric values into millions$
def zeros_to_millions(value):
  if isinstance(value,(int,float)):
    return value/1000000  # 1 million dollars = 1000 thousands
  else:
    return value

# Function to format column names
def format_column_name(column_name):
    result = []
    for char in column_name:
        if char.isupper() and result:
            result.append(' ')
        result.append(char)
    return ''.join(result).title()

# Skip the first two rows (containing date and currency info)
numeric_ann_df = annual_df.iloc[:,2:].copy()

# Convert specific columns to numeric data types (e.g., columns 0, 1, 2, 3, 4)
column_indices = list(range(len(numeric_ann_df.columns))) # Adjust these column indices as needed

#converting object data type into numberic value so the 000s can be converted into millions
for ann_idx in column_indices:
    numeric_ann_df[numeric_ann_df.columns[ann_idx]] = pd.to_numeric(numeric_ann_df[numeric_ann_df.columns[ann_idx]], errors='coerce')

# Now, you can proceed with converting the numeric values to millions
numeric_ann_df = numeric_ann_df.applymap(zeros_to_millions)

#Select the first two columns of the original income statement
header_columns=annual_df.iloc[:,:2]

#combine both; the header columns that contain the date and currency type, and the numeric values converted into millions
combined_ann=pd.concat([header_columns,numeric_ann_df],axis=1).fillna('-')

# Apply the formatting function to each column name
combined_ann.columns = [format_column_name(col) for col in combined_ann.columns]



                                                                    ########## Annual Reports Code #############
#extracting annual reports from the json response file
quarterly_reports=is_data['quarterlyReports']

#converting the json file into a Pandas DataFrame
quarterly_df=pd.DataFrame(quarterly_reports)

# Skip the first two rows (containing date and currency info)
numeric_quar_df = quarterly_df.iloc[:,2:].copy()

# Convert specific columns to numeric data types (e.g., columns 0, 1, 2, 3, 4)
quarterly_indices = list(range(len(numeric_quar_df.columns))) # Adjust these column indices as needed

#converting object data type into numberic value so the 000s can be converted into millions
for quar_idx in quarterly_indices:
    numeric_quar_df[numeric_quar_df.columns[quar_idx]] = pd.to_numeric(numeric_quar_df[numeric_quar_df.columns[quar_idx]], errors='coerce')

# Now, you can proceed with converting the numeric values to millions
numeric_quar_df = numeric_quar_df.applymap(zeros_to_millions)

#Select the first two columns of the original income statement
header_quar=quarterly_df.iloc[:,:2]

#combine both; the header columns that contain the date and currency type, and the numeric values converted into millions
combined_quar=pd.concat([header_quar,numeric_quar_df],axis=1).fillna('-')
combined_quar.columns = [format_column_name(col) for col in combined_quar.columns]

combined_quar_T=combined_quar.T
combined_quar_T
```
